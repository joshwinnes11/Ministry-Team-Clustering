### Cleaning 2021 Ministry Teams

```python
df_21_MT.columns
```

```python
columns = ['First Name', 'Last Name', 'What ministry are you a part of?', 'What is your position on the cabinet?','What other student leadership role do you have?']
df_21_MT = df_21_MT[columns]
df_21_MT.isna().sum()
```

### Cleaning 2021 Orgs

```python
df_21_Org.columns
```

```python
df_21_Org.head()
```

```python
org_columns = ['First Name', 'Last Name', 'I am a....1']
df_21_Org = df_21_Org[org_columns]
df_21_Org.isna().sum()
```

```python
df_21_Org
```

```python
org_cabs_dict = {'Ministry Trip Cabinet Member or Student Evangelism Cabinet Member':'Member',
                 'Ministry Trip Cabinet Director or Student Evangelism Director': 'Director'}

df_21_Org['I am a....1'] = df_21_Org['I am a....1'].map(org_cabs_dict)
df_21_Org
```

```python
columns_correct_order = ['First', 'Last', 'Ministry Team', 'Role', 'Other Leadership Roles']
```

```python
df_21_Org['Other Leadership Roles'] = 'None'
```

```python
df_21_MT = df_21_MT[columns_correct_order]
df_21_Org = df_21_Org[columns_correct_order]
```

```python
df_21 = pd.concat([df_21_Org,df_21_MT],ignore_index=True)
```
